```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv("TCGA/dataset.csv", index_col=0)
outcome = pd.read_csv("TCGA/outcome.csv", index_col=0)
```

```python
combined = dataset.join(outcome)
```

```python
#columns to delete, which columns have missing values
delcol = combined.isna().sum()[combined.isna().sum() > 0]
delcol
```

```python
#rows to delete, rows with missing values
delrow = combined.isna().sum(axis=1)[combined.isna().sum(axis=1) > 0]
delrow
```

```python
#clean combined dataset by dropping columns and rows with missing values
combined_clean = combined.drop(columns=delcol.index, index=delrow.index)
combined_clean.shape
```
