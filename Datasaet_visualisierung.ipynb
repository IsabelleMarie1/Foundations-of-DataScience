```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
os.getcwd()
os.listdir()
```

```python
dataset = pd.read_csv("TCGA/dataset.csv", index_col=0) # Note the use of a relative path!
outcome = pd.read_csv("TCGA/outcome.csv", index_col=0) # Note the use of a relative path!
```

```python
outcome.head()
```

```python
sns.histplot(outcome["BRCA_subtype"], kde=True)
```

```python
print(dataset.shape)
print(dataset.columns)
print(dataset.head())
print(dataset.info())
```

```python
combined=dataset.join(outcome)
print(combined.head())
print(combined.info())
```

```python
combined = combined.dropna(subset=["BRCA_subtype"])
combined["BRCA_subtype"].value_counts()
```

```python
combined.groupby("BRCA_subtype").mean().T
```

```python
combined.groupby("BRCA_subtype").mean().T.var(axis=1).sort_values(ascending=False).head(10)
```

```python
col = "AGR3"

sns.boxplot(x="BRCA_subtype", y=col, data=combined)
plt.title(col)
plt.show()
```
